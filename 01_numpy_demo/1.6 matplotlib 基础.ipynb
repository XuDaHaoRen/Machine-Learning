### matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 定义一个数组 （0,100，分成 100份）
x = np.linspace(0,10,100)
```

```python
# 对 x 中的每个值求 sin
siny = np.sin(x)
# 绘制多条曲线 显示 cons 的关系
cosy = np.cos(x)
tany = np.tan(x)
```

### 1.绘制折线图

```python
# 创建一个坐标,并显示。老的版本需要调用 plt.show
# plot 方法两个参数分别对应 x 轴的取值，y 轴的取值
# 这个方法可以添加多条曲线，并绘制颜色
#linestyle 线条样式
#其实这个是绘制的折线图，但是因为数据比较多给人的感觉是曲线图
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle='dashed')
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red")
# 对 plot 的 x,y轴 再描述一个范围
plt.xlim(-5,15)
plt.ylim(0,1)
```

```python
#仅用一个方法 对 x,y轴设置范围,里面需要传入一个数组。
#数组的前两个值是 x 轴的范围，后两个值是y 轴的范围
plt.plot(x,siny)
plt.plot(x,cosy,color="red")
plt.axis([-1,11,-2,2])
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red")
#为坐标轴添加字符串说明
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
# 为每个线添加注释
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color="red",label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
#为每个线添加注释的时候一定加这句话
plt.legend()
# 对图示添加标题
plt.title("This is too difficult for me")
```

### 2.Scatter Plot
绘制散点图，里面的样式设置与折线图是一模一样的

```python
plt.scatter(x,siny)
```

```python
# 绘制正态分布.这个正态分布可以看出来，中间密集，四周稀疏
x=np.random.normal(0,1,10000)
y=np.random.normal(0,1,10000)
#alpha 透明度
plt.scatter(x,y,alpha=0.1)
```
