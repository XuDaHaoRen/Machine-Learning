### 多项式回归

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# 生成数据
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
# y 呈一种多项式的关系
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, 100)
```

```python
# 显示数据
plt.scatter(x, y)
plt.show()
```

### 线性关系拟合数据

```python
from sklearn.linear_model import LinearRegression
#将 X，y 数据进行拟合
lin_reg =  LinearRegression()
lin_reg.fit(X,y)
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
# 将拟合的结果进行显示，拟合的是一个线性关系
#相比于下面这里是一个线性关系，画出来的 y 不可能是折线，所以这里不用排序
plt.plot(x, y_predict, color='r')
plt.show()
```

### 添加一个特征,形成多项式的关系

```python
X2 = np.hstack([X,X**2])
```

```python
X2.shape
```

```python
#将这个X2进行拟合
lin_reg2 =  LinearRegression()
lin_reg2.fit(X2,y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
# 这样实现的线是非常混乱的折线
plt.plot(x, y_predict2, color='r')
plt.show()
```

```python
# 由于 x 中的数据是乱的，所以会产生上面的折线，所以这次对数据进行排序
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')
```

```python
# X 前面的系数是1，X^2 前面的系数是0.5
lin_reg2.coef_
```

```python
#拮据是 2
lin_reg2.intercept_
```
